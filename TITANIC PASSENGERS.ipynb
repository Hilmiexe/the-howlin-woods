```python
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 })
```

```python
plot_correlation_map(new_data)
#this function is used in order to describe the correlation between multiple factors and the impact of one on the another .
```

```python
#Extraction of the titles from the names of the passengers and the creation of a new column
def Extract(ch) :
    T=ch.split(', ')[1].split('. ')[0]
    return T
new_data['Titles']=new_data['Name'].map(Extract)
new_data
```

```python
new_data['Titles'].value_counts()
```

```python
grid=sns.FacetGrid(new_data,row='Titles',col='Survived',size=2.2,aspect=1.6)
grid.map(sns.barplot,'Age',alpha=0.6,ci=None)
grid.add_legend()
```

```python
Title_Dictionary = {"Capt":"Officer",
                    
                    "Col":"Officer",
                    
                    "Major":"Officer",
                    
                    "Dr":"Officer",
                    
                    "Rev":"Officer",

                    "Jonkheer":"Royalty",

                    "Don":"Royalty",

                    "Sir" :"Royalty",

                   "Lady" :"Royalty",

                  "the Countess":"Royalty",

                    "Dona":"Royalty",

                    "Mme":"Miss",

                    "Mlle":"Miss",

                    "Miss" :"Miss",

                    "Ms":"Mrs",

                    "Mr" :"Mrs",

                    "Mrs" :"Mrs",

                    "Master" :"Master"}
```

```python
new_data['Titles']=new_data['Titles'].map(Title_Dictionary)
new_data['Titles'].value_counts()
sns.countplot(x='Titles',data=new_data)
plt.title("Passenger's Titles")
```
