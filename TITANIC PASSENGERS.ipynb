```python
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 })
```

```python
plot_correlation_map(new_data)
#this function is used in order to describe the correlation between multiple factors and the impact of one on the another .
```

# Applying logistic regression model


```python
#import relevant libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

#features extraction
x=new_data[["Age", "Pclass", "Sex", "Embarked"]]
y=new_data["Survived"]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.35,random_state=40) #splitting data with test size of 35%

logreg = LogisticRegression()   #build our logistic model
logreg.fit(x_train, y_train)  #fitting training data
y_pred  = logreg.predict(x_test)    #testing model’s performance
print("Accuracy={:.2f}".format(logreg.score(x_test, y_test)))
```

```python
confusion_matrix =pd.crosstab(y_test,y_pred,rownames=['Actual'], colnames=['Predicted'])
confusion_matrix
```

```python
##Classification report
from sklearn.metrics import classification_report 
print(classification_report(y_test,y_pred))
```

```python
#predict probabilities 
pred_prob=logreg.predict_proba(x_test)
```

```python

from sklearn.metrics import roc_curve 
#Roc curve for the binary classifier 
fpr, tpr, thresh=roc_curve(y_test,pred_prob[:,1], pos_label=1)
#roc curve for fpr=tpr
random_probs=[0 for i in range(len(y_test))]
p_fpr, p_tpr, _ = roc_curve(y_test, random_probs, pos_label=1)
```

```python
p_tpr
```

```python
#computing the model roc_score to figure out its performance
from sklearn.metrics import roc_auc_score

# auc scores
auc_score = roc_auc_score(y_test, pred_prob[:,1])
print(auc_score)
```

```python
import seaborn as sns
#plot the roc curve
plt.style.use('seaborn')
#plot roc curves 
plt.plot(fpr, tpr, linestyle='--', color='green', label='logisitic regression')
plt.plot(p_fpr, p_tpr, linestyle='--', color='blue')
# title
plt.title('ROC curve')
# x label
plt.xlabel('False Positive Rate')
# y label
plt.ylabel('True Positive rate')

plt.legend(loc='best')
plt.savefig('ROC',dpi=300)
plt.show()
```

# Using the ROC/AUC validiation matrix

## Explaining the ROC/AUC validation matrix for binary classification models

```python
#An ROC curve (receiver operating characteristic curve) is a graph showing the performance of a classification model at all classification thresholds. This curve plots two parameters:
#True Positive Rateand False Positive Rate
#The higher the AUC, the better the performance of the model at distinguishing between the positive and negative classes.
#AUC stands for "Area under the ROC Curve." That is, AUC measures the entire two-dimensional area underneath the entire ROC curve (think integral calculus) from (0,0) to (1,1).
```
