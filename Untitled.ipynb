```python
def diviseur(y,a,b) :
    mylist=[]
    for x in range(a,b):
          if x%y==0 :
                mylist.append(x)
    return mylist
l = diviseur(5,25,50)
print(l)
import pandas as pd
new_data = pd.DataFrame(l,columns=["diviseur"])
new_data.rename(columns={0: 'diviseur'})
new_data
```
